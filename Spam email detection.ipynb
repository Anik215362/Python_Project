# Import Library & Read dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df.drop_duplicates(inplace=True)
```

# Separate X & Y for dependent & Independent value

```python
x = df.text.values
```

```python
y = df.spam.values #Dependent
```

# split dataset

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.2)
```

# Data preprocessing

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
x_train = cv.fit_transform(xtrain)
x_train.toarray()
```

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train,ytrain)

x_test = cv.transform(xtest)
```

```python
x_test.toarray()
```

```python
model.score(x_test,ytest)
```

```python
emails = ['hey i am looking for a car driver','I will pay him after work as free giveaway as servey']
```

```python
cv_emails = cv.transform(emails)
```

```python
model.predict(cv_emails)
```

```python
pred = model.predict(x_test)

pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,pred)
```

# Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier_rf = RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5, n_estimators=100, oob_score=True)
```

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
```

```python
jalal = {
    'max_depth': [2,3,5,10,20],
    'min_samples_leaf': [5,10,20,50,100,200],
    'n_estimators': [10,25,30,50,100,200]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=rf,
                           param_grid=jalal,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")
```

```python
grid_search.fit(x_train, ytrain)
```

```python
grid_search.best_score_
```
